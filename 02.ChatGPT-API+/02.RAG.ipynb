## Few-shot CoT

몇 개의 예시에서 추론 패턴을 보여주면, 모델이 그 패턴을 학습해서 새로운 문제에 적용

```python
# 수학 문제 Few-Shot CoT
few_shot_math = """
다음은 백분율 문제를 단계별로 해결하는 예시입니다:

예시 1:
문제: 200명 중 30%가 남학생이면 남학생은 몇 명?
풀이:
1. 전체 인원: 200명
2. 남학생 비율: 30% = 0.3  
3. 남학생 수: 200 × 0.3 = 60명
답: 60명

예시 2:
문제: 80점 만점에서 64점을 받았다면 몇 %?
풀이:
1. 만점: 80점
2. 받은 점수: 64점
3. 백분율: (64 ÷ 80) × 100 = 0.8 × 100 = 80%
답: 80%

이제 다음 문제를 같은 방식으로 해결해주세요:
문제: 150명 중 여학생이 45%라면 남학생은 몇 명?
"""

# 예상 응답:
"""
풀이:
1. 전체 인원: 150명
2. 여학생 비율: 45%
3. 남학생 비율: 100% - 45% = 55%
4. 남학생 수: 150 × 0.55 = 82.5명 → 83명
답: 83명
"""
```
